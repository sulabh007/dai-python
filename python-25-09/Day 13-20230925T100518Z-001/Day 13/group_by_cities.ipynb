```python
y=g['temperature'].max()
y
cities=y.index
values=y.values
cols=['g','b','k']

import matplotlib.pyplot as plt
plt.pie(values,labels=cities,colors=cols, startangle=90,explode=(0,0.2,0)) #,counterclock=False)
plt.show()
```
